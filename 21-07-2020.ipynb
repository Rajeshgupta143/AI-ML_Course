```python
import imutils
```

```python
face_cascade= cv2.CascadeClassifier('C:/Users/lenovo/Desktop/summer internship/face data/haarcascade_frontalface_default.xml')
eye_cascade= cv2.CascadeClassifier('C:/Users/lenovo/Desktop/summer internship/face data/haarcascade_eye.xml')
smile_cascade= cv2.CascadeClassifier('C:/Users/lenovo/Desktop/summer internship/face data/haarcascade_smile.xml')
```

```python
## to read the image in array
%matplotlib auto
img=cv2.imread('face.jpg')
print(img)
## resize the image
img=imutils.resize(img,width=1200)
## converting BGR image into GrayScale
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2.imshow('gray color',gray)
```

```python
plt.imshow(gray)
```

```python
## use face cascade to extract the face cordinates
#1.3- Threshold,5 - number of neighbouring pixels
faces=face_cascade.detectMultiScale(gray,1.3,5)
#print no of faces
print(len(faces))
```

```python
# print the value
print(faces) ## (X,Y,W,H)
```

```python
## to read the image in array
%matplotlib auto
img=cv2.imread('faces.jpg')
#print(img)
## resize the image
img=imutils.resize(img,width=1200)
## converting BGR image into GrayScale
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2.imshow('gray color',gray)
```

```python
plt.imshow(gray)
```

```python
## use face cascade to extract the face cordinates
#1.3- Threshold,5 - number of neighbouring pixels
faces=face_cascade.detectMultiScale(gray,1.05,3)
#print no of faces
print(len(faces))
```

```python
# print the value
print(faces) ## (X,Y,W,H)
```

```python

for (x,y,w,h) in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(255,255,255),1)
    roi_gray=gray[y:y+h,x:x+w]
    roi_color=img[y:y+h,x:x+w]
    
```

```python
plt.imshow(img)
```

```python
%matplotlib auto
cv2.imshow('predicted_faces',img)
```

```python
%matplotlib auto
cv2.imshow('predicted faces',roi_color)
```

```python
# image with predicted faces stored in to predicted.jpg
cv2.imwrite("predicted.jpg",img)
```

```python
smile=smile_cascade.detectMultiScale(roi_gray,2.5,10)
for(sx,sy,sw,sh) in smile:
    cv2.rectangle(roi_color,(sx,sy),(sx+sw,sy+sh),(0,255,255),2)
eyes= eye_cascade.detectMultiScale(roi_gray,2,7)
for (ex,ey,ew,eh) in eyes:
    cv2.rectangle(roi_color)
```
