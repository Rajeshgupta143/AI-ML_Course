```python
# importing 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Loading the dataset from sklearn library
from sklearn.datasets import load_boston # importing the boston data from sklearn
boston=load_boston()
```

```python
boston_df=pd.DataFrame(boston.data,columns=boston.feature_names)
boston_df.head()
```

```python
boston_df['MEDV']=boston.target
boston_df.head()
```

```python
## To find the correlation we use corr()
boston_df.corr()
```

```python
## Visualize the correlation

sns.heatmap(boston_df.corr())
```

```python
## Visualize the correlation
plt.figure(figsize=(14,14)) # increse 
sns.heatmap(boston_df.corr(),annot=True)
```

```python
## Visualize the correlation
plt.figure(figsize=(14,14))
sns.heatmap(boston_df.corr(),annot=True,cmap='YlGnBu')
```

```python
corralation=boston_df.corr()
corr=corralation[corralation>0.7]
corr
```

```python
corralation=boston_df.corr()
corr=corralation[corralation>0.7]
plt.figure(figsize=(13,13))
sns.heatmap(corr,cmap='Greens',annot=True)
```

```python
X=boston_df.drop(['MEDV'],axis=1)
y=boston_df.MEDV
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test,y_train, y_test = train_test_split(X, y, test_size=0.25,
                                               random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
lm.fit(X_train,y_train)
```

```python
print(lm.intercept_)
lm.coef_
```

```python
coefficients=pd.DataFrame([X_train.columns,lm.coef_]).T
coefficients
```

```python
# y=b0+b1x1+b2x2+.......b13x13
# y=36.93325545711978+(-0.117735)
```

```python
y_train_pred=lm.predict(X_train)
y_train_pred
```

```python
from sklearn.metrics import r2_score
print('R^2:',r2_score(y_train,y_train_pred))
```

```python
from sklearn.metrics import r2_score
print('Adjusted R^2',1-(1-r2_score(y_train,y_train_pred))*(len(X_train)-1)/
                           (len(X_train)-X_train.shape[1]-1))
```

```python
# check the error betwenn our Actual value and the predicted value
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
print('MAE:', mean_absolute_error(y_train, y_train_pred))

print('MSE:', mean_squared_error(y_train, y_train_pred))

print('RMSE', np.sqrt(mean_squared_error(y_train, y_train_pred)))
```

```python
#2nd way:
numerator = (1-r2_score(y_train_pred,y_train))*(len(X_train)-1)
denominator = (len(X_train)-X_train.shape[1]-1)

print("Adjusted R^2",1- (numerator/denominator))
```
