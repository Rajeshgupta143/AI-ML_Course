```python
import numpy as np
import pandas as pd
import sklearn
```

```python
data=pd.read_csv("https://raw.githubusercontent.com/Mounika-Kajjam/Datasets/master/weather_data.csv")
data
```

```python
data=data.replace({'temperature':-99999,'windspeed':-99999,'event':'0'},np.nan)
data
```

### interpolate() : Based on the column values interpolate function will come up with the guess value to be filled Nan

```python
## interpolate fill the nan value automatic based on guess
int_data= data.interpolate()
int_data
```

### dropna(): Drops the column or rows which have missing value

```python
drop_data=data.dropna()
drop_data
```

```python
drop_data=data.dropna(how='any')
drop_data
```

```python
### drop the column having any value
drop_data=data.dropna(axis=1) #1 means row
drop_data
```

```python
# if any missing value in column it will drop
rop_data=data.dropna(axis=0) # 0 means column
rop_data
```

```python
# if missing value present in all row then it will drop
drop_all_data=data.dropna(how='all')
drop_all_data
```

### Handling Missing Values using Scikit-Learn

```python
### SimpleImputer -imputer only work on integer
#- it use 2 arguments
#- 1 -- which value to be replaced
#- 2 -- what value it has to be replaced(strategy =)
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan,strategy='mean')
```

```python
## it is a 2 step transformation
#-fit
#-transform
```

```python
imputer.fit(data.iloc[:,1:3]) # indices of column 1 and column 2
```

```python
data.iloc[:,1:3] = imputer.transform(data.iloc[:,1:3])
data
```

```python
price=pd.read_csv("https://raw.githubusercontent.com/Mounika-Kajjam/Datasets/master/homeprices.csv")
price
```

```python
### Accessing town column from price dataframe using [] notation
price['town']
```

```python
price.town # Accessing town column from the price dataframe using dot notation
```

```python
## Covert the categorical data into numerical columns using get_dummies
dummy_set=pd.get_dummies(price.town)
dummy_set
```

```python
#### Concatenating the original dataframe and dataframe of dummy columns
merged_data= pd.concat([price,dummy_set],axis=1)
merged_data
```

```python
# Creating dummy columns and dropping the dummy column
pd.get_dummies(price,columns=['town'],drop_first=True)
```

### Dummy Variable Trap:
    -The Dummy Variable Trap is a scenario in which the independent variables are multicollinear - a scenario in which two or more variables are highly correlated; in simple terms one variable can be predicted from the others.
    - We need to drop the original town column as we have the dummy columns. Also we need to drop one of the dummy variables column in order to avoid the multi-collinearity, which creates the problem of dummy variable trap.  
    - Hence, one dummy variable is highly correlated with other dummy variables. Using all dummy variables for regression models lead to dummy variable trap.And the dummy variable trap mess up the machine learning model.
